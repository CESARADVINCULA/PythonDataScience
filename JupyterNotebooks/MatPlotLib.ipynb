```python
#BAR CHART

import numpy as np
import matplotlib.pyplot as plt

objects = ('Python','C++','Java','Perl','Scala','Lisp')
y_pos=np.arange(len(objects))
performance= [10,11,6,4,2,1]

plt.bar(y_pos, performance, align='center', alpha=0.5) 
#align : {'center', 'edge'}, optional, default: 'center'
#alpha :  Intensity the color of the bar

plt.xticks(y_pos, objects) #y_pos = N bars 
plt.ylabel('Usage')
plt.title('Programming language plotting by MatPlotLib')

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 3.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
print(s)
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

fig.savefig("testABC.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

figParent = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
#figParent.savefig("parentFig.png")

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

```python
#Heatmap Chart

import numpy as np
import numpy.random
import matplotlib.pyplot as plt

#Create data
x = np.random.randn(4096)
y = np.random.randn(4096)


#Create heatmap
heatmap, xedges, yedges = np.histogram2d( x, y, bins=(64,64) )
extent = [ xedges[0], xedges[-1], yedges[0], yedges[-1] ]

#Plot Heatmap
plt.clf()
plt.title('Python heatmap example')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.imshow(heatmap , extent = extent)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (30,10))

plt.subplot2grid((2,3),(0,0))
data.Survived.value_counts().plot(kind='bar', alpha = 0.5)
plt.title('Sobrevivieron - Cantidad Total')

plt.subplot2grid((2,3), (0,1))
data.Survived.value_counts(normalize = True).plot(kind='bar', alpha =0.5)
plt.title('Sobrevivieron - Porcentaje Total')


plt.show()
```
