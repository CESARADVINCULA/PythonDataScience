```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# url that we are scraping
url = "http://www.basketball-reference.com/draft/NBA_2014.html"

# this is the html from the given url
html = urlopen(url)
print(html)
```

```python
soup = BeautifulSoup(html)
print(soup)
```

```python
type(soup)  # we see that soup is a BeautifulSoup object
```

```python
data_rows = soup.findAll('tr')[2:]  # skip the first, point out on 2 header rows
type(data_rows)
print(data_rows)
```

```python
player_data = [[td.getText() for td in data_rows[i].findAll('td')]
            for i in range(len(data_rows))]
print(type(player_data))

print(len(player_data))

print(player_data)
```

```python
column_headers = [th.getText() for th in 
                  soup.findAll('tr', limit=2)[1].findAll('th')]
print(type(column_headers))
print(column_headers)
```

```python
column_headers=['PK','Tm','Player','College','Yrs','G','MP','PTS','TRB','AST','FG%','3P%','FT%','MP','PTS','TRB','AST','WS','WS/48','BPM','VORP']
print(type(column_headers))
column_headersdf = pd.DataFrame(player_data, columns=column_headers)
```
